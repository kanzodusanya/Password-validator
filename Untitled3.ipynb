```python
# AI Decision Judge 1.0
print("Welcome to Decision AI")


decision = input("What is your decision?" )

good_list = ["help", "share", "study"]
bad_list = ["lie", "steal", "fight"]


if decision is good_list:
    print("AI Verdict: That is a RIGHT decision!")


elif decision is bad_list:
    print("AI Verdict: That is a WRONG decision.")

else:
    print("AI Verdict: I need more info.")
```
