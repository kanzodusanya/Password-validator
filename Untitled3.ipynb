```python
# AI Decision Judge 1.0
print("Welcome to Decision AI")


decision = input("What is your decision?" )

good_list = ["help", "share", "study"]
bad_list = ["lie", "steal", "fight"]


if decision is good_list:
    print("AI Verdict: That is a RIGHT decision!")


elif decision is bad_list:
    print("AI Verdict: That is a WRONG decision.")

else:
    print("AI Verdict: I need more info.")
```

```python
print("Bill Split Calculator")

bill_amount = float(input())
tip_percentage = float(input())
num_people = int(input())

tip_amount = (tip_percentage / 100) * bill_amount
total_amount = bill_amount + tip_amount
amount_per_person = total_amount / num_people

print(f"Total (including tip): ${total_amount}")
print(f"Each person pays: ${amount_per_person}")
```
